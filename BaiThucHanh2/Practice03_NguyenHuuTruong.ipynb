```python
#bai2

#a
lazada = pd.DataFrame(pd.read_json("LazadaProductData.json"))
lazada
```

```python
#b
lazada.loc[lazada["p_cate"] == "Máy vi tính & Laptop"]

lazada.loc[lazada["p_rating"] > 4.0]
```

```python
#c
lazada["p_cate"].value_counts()

lazada.sort_values("p_rating", ascending=False).head(10)

lazada.sort_values("p_rating").head(100)
```

```python
#d
lazada["p_price"] = lazada["p_price"].str.replace("₫", "").str.replace(".", "")
lazada["p_price"] = pd.to_numeric(lazada["p_price"])

lazada["p_rate1star"] = lazada["p_rate1star"].fillna(0) #dien toan bo gia tri NaN thanh 0
lazada["p_rate2star"] = lazada["p_rate2star"].fillna(0)
lazada["p_rate3star"] = lazada["p_rate3star"].fillna(0)
lazada["p_rate4star"] = lazada["p_rate4star"].fillna(0)
lazada["p_rate5star"] = lazada["p_rate5star"].fillna(0)

lazada.to_json("LazadaProducts_Cleaned.json")
```
